```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 1.5
v0 = .1
E = 90
def V1(x,v0,a):

    ret = -v0 * ((a-np.abs(x))/a)**3
    ret[np.abs(x)>a]=0
    return ret
def p1(x,v0,E,a,b):
    return np.sqrt(E-V1(x,v0,a))
x = np.linspace(-b,b,100)
E_list = [0.00000000001,0.0001]
for E in E_list:
    plt.plot(x,p1(x,v0,E,a,b))
    plt.plot(x,-p1(x,v0,E,a,b))
```
