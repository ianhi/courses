```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.style.use(['figsize'])
```

# Prob 3.c

Assuming J = 1

```python
V = lambda m, beta: (m**2/2-np.log(np.cosh(4*m*beta))/(4*beta))/beta
m = np.linspace(-2,2,100)

for beta in [.1,.25,.5]:
    plt.plot(m,V(m,beta),label=r'$\beta = {:}$'.format(beta))
plt.axhline(color='black')
plt.legend()
plt.show()
```

This is very similar to figure 9.22!! They display similar behavior with values of a parameter near a critical point. Having one stable state that decays into a metastable state flanked by two stable states.

# 3. c

```python
Vc = lambda m, beta, H: (m**2/2-np.log(np.cosh((H+4*m)*beta))/(4*beta))/beta
m = np.linspace(-2,2,100)
beta = .5
for H in [0,.5,1,1.5,1.6]:
    plt.plot(m,Vc(m,beta,H),label=r'$H = {:}$'.format(H))
plt.axhline(color='black')
plt.legend()
plt.show()
```

This has the same basic shape as 11.2a, though here there is for some values of the parameters a stable state to left of the metastable state.

## Finding critical H to lose metastable state.
To find the critical value at which the metastable state becomes totally unstable I will plot the first derivative and search for at what H value there ceases to be a zero in the first derivative at the position of the metastable sol'n

```python
m = np.linspace(-.9,-.5,100)
for H in np.linspace(1.,1.1,10):
    plt.plot(m,np.gradient(Vc(m,beta,H)),label=r'$H = {:}$'.format(H))
plt.axhline(color='black')
plt.legend()
plt.show()
```

### I find H ~= 1.06 to be the value where the metastable state becomes completely unstable.
