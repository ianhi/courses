```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use(['fivethirtyeight', 'figsize'])
_SAVEDIR_ = 'Tex/figures/'
from IPython.display import HTML


def prog_bar(cur_val,max_val,msg = ""):
    import sys
    sys.stdout.write('\r')
    eq = int(np.ceil(np.true_divide(cur_val*100,max_val*5)))
    sys.stdout.write("[{:20s}] {}/{} {:}  ".format('='*eq, cur_val+1,max_val,msg))
    sys.stdout.flush()
```

```python
# Variables

N_realizations = 9
a0 = 30

gamma  = 1.0
a0     = 30.0         #! external field  
wk0    = 0.5          #! spatial periodicity
b      = 0.3         # ! exponential scale
gamma  = 1.0    
vth    = 0.1
rnl    = 0.0#          ! non linear damping rate
dt    = 0.01
nout  = 50
Nsteps = 300
v_record = np.zeros([int(N_realizations*Nsteps/10)])
x_record = np.zeros([int(N_realizations*Nsteps/10)])
x0 = np.array([-10.,-5.0,-3.0,-1.0,0.0 ,1.0,3.0 ,5.0 ,10. ])

n_out = 50
count = 0
for i in range(N_realizations):
    
# initial conditions
        x = x0[i]
        v = 0.0
# time evolution
        vrave=0.
        vrvar=0.
        for it in range(Nsteps):
            rv = (2.0*np.random.rand()-1.0)
# single well, linear
            if(rnl == 0):
                z  = gamma*dt
                gd = np.exp(-z)
                gc = (1.0-np.exp(-z))/z
                gr = np.sqrt(0.5*(1.0-np.exp(-2.*z))/z)
# gamma-integrator
                xk = 6.28*wk0*x
                a  = -a0*np.sin(xk)*np.cos(xk)*np.exp(b*np.abs(x))
                vc = a*dt
                vr = rv*vth*(np.sqrt(3.0))

                vrave = vrave + vr
                vrvar=vrvar+vr*vr
                vnew = gd*v + gc*vc + gr*vr
            else:
# double well: v=0 unstable, v=\pm 1 stable
# plain euler because we don't have the analytical solution of the purely damped problem 
                z = gamma*dt
# note the sign change because v=0 must be unstable
                xk = 6.28*k0*x
                a  = -a0*sin(xk)*cos(xk)*exp(b*x)
                vnew = v*(1.0+z*(1.0-rnl*v*v)) + a*dt + vth*rv*sqrt(dt)
            xnew = x + 0.5*(v+vnew)*dt

            if it%n_out ==0:
                v_record[count] = vnew
                x_record[count] = xnew
                count += 1
            x = xnew
            v = vnew
                
```

```python
# Lets make the potential:
x1 = 2
x2 = 5
x= np.linspace(-5,10,1000)
# plt.plot(x,-np.sinc((x-x1))-np.sinc(x-x2))
def sinc_prime(a):
    return np.nan_to_num((a*np.cos(a)-np.sin(a))/a**2)
def pot_deriv(x):
    return -sinc_prime((x-x1)*np.pi)-sinc_prime((x-x2)*np.pi)
def pot(x):
    return -np.sinc((x-x1))-np.sinc(x-x2)
plt.plot(x,pot(x),label='U(x)')
plt.plot(x,pot_deriv(x),label=r'$\partial_x U$')
plt.xlabel('X')
plt.tight_layout()
plt.legend()
plt.savefig(_SAVEDIR_+"lange_potential.png")
# plt.show()
```
